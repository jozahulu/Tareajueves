```python
theta_1 = np.random.random(1000)*2.*pi
r2 = np.random.random(1000)
rn = r2**(0.5)
cosphi = np.random.uniform(-1,1,1000)
phi = np.arccos(cosphi)
```

```python
x1 = r2*cos(theta_1)*sin(phi)
y1 = r2*sin(theta_1)*sin(phi)
z1 = r2*cosphi
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x1, y1, z1)
```

```python
x2 = rn*cos(theta_1)*sin(phi)
y2 = rn*sin(theta_1)*sin(phi)
z2 = rn*cosphi
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x2, y2, z2)
```
